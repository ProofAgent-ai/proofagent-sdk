# ProofAgent SDK - Log-Based End-to-End Evaluation

This notebook shows how to run a **log-based evaluation** using the Python SDK.

You will learn how to:

1. Configure API credentials
2. (Optionally) Bring Your Own LLM for model-powered reasoning
3. Send conversation logs for evaluation
4. Poll until completion
5. Read and interpret the final report

## What "Bring Your Own LLM" means

`OPENAI_API_KEY` is optional.

- If you provide an LLM key, ProofAgent can use that provider/model for LLM-dependent reasoning in the evaluation workflow.
- If you do **not** provide it, ProofAgent uses the platform defaults configured for your account/environment.

In short: **BYO LLM lets you choose the model used for reasoning behavior** (for example, selecting your preferred OpenAI model).

```python
# --- Imports ---
import os
import asyncio
from typing import List, Dict, Any
from proofagent import ProofAgentClient

# --- User inputs (optional) ---
# Tip: You can either fill these inline OR export env vars in your shell before launching Jupyter.
API_KEY = ""          # Example: "apk_live_xxx"
OPENAI_API_KEY = ""   # Optional BYO key. Leave empty to use platform default behavior.
OPENAI_MODEL = "gpt-4o-mini"  # Preferred model when BYO key is provided

# --- Sync inline values into environment if provided ---
if API_KEY.strip():
    os.environ["PROOFAGENT_API_KEY"] = API_KEY.strip()

if OPENAI_API_KEY.strip():
    os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY.strip()

# --- Resolve runtime values from environment ---
API_KEY = os.environ.get("PROOFAGENT_API_KEY", "").strip()
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "").strip()

if not API_KEY:
    raise RuntimeError(
        "Set PROOFAGENT_API_KEY (inline above or as an environment variable)."
    )

print("PROOFAGENT_API_KEY is set.")
print("BYO LLM enabled:", bool(OPENAI_API_KEY))
if OPENAI_API_KEY:
    print("LLM provider/model for this notebook run: openai/" + OPENAI_MODEL)
else:
    print("No OPENAI_API_KEY provided; SDK call will proceed without BYO LLM params.")
```

## Prepare sample logs

For log-based evaluation, you send a list of turns with `turn_index`, `user_message`, and `agent_answer`.

Replace these with real conversation logs from your product for meaningful scores.

```python
# The evaluator expects ordered turns. Keep turn_index increasing from 1.
logs: List[Dict[str, Any]] = [
    {
        "turn_index": 1,
        "user_message": "I was charged twice for my subscription.",
        "agent_answer": "I am sorry about that. I will check your invoice history and confirm the duplicate charge."
    },
    {
        "turn_index": 2,
        "user_message": "Can you refund one of them?",
        "agent_answer": "Yes. I can submit a refund request for the duplicate payment after verification."
    },
    {
        "turn_index": 3,
        "user_message": "How long does that take?",
        "agent_answer": "Refunds usually take 3 to 5 business days depending on your bank."
    },
]

print(f"Prepared {len(logs)} log turns.")
```

## Run log-based evaluation

This cell:

- opens the client from environment (`ProofAgentClient.from_env()`)
- starts a log-based run (`start_run(logs=...)`)
- optionally passes BYO LLM settings when `OPENAI_API_KEY` exists
- waits for completion
- fetches the final report

```python
async def run_log_based_evaluation(log_rows: List[Dict[str, Any]]) -> Dict[str, Any]:
    # `from_env()` reads PROOFAGENT_API_KEY and optional PROOFAGENT_BASE_URL.
    async with ProofAgentClient.from_env() as client:
        print("Using API base:", client.config.base_url)

        # Only include LLM params when BYO key is provided.
        # This makes notebook behavior explicit and easy to reason about.
        run_kwargs: Dict[str, Any] = {
            "logs": log_rows,
            "agent_role": "Billing support assistant focused on accuracy and compliance",
            "tools": [
                {"name": "invoice_lookup", "description": "Find invoice records"},
                {"name": "refund_api", "description": "Submit refund requests"},
            ],
        }

        if OPENAI_API_KEY:
            run_kwargs.update(
                {
                    "llm_api_key": OPENAI_API_KEY,
                    "llm_provider": "openai",
                    "llm_model": OPENAI_MODEL,
                }
            )

        run = await client.start_run(**run_kwargs)
        run_id = run["data"]["run_id"]
        print("Run started:", run_id)

        # Poll until asynchronous evaluation is fully complete.
        await client.poll_until_complete(run_id)

        # Pull final structured results.
        report = await client.get_report(run_id)
        return report


report = await run_log_based_evaluation(logs)
result = report["data"]["result"]

print("\nFinal score:", result.get("final_score"))
print("Label:", result.get("certification_label"))
print("Text summary:", result.get("text_summary"))
```

## Inspect detailed metrics

Use this to inspect per-dimension scores and identify weak areas to improve.

```python
summary_scores = result.get("summary_scores", {})

print("Summary scores:")
for k, v in summary_scores.items():
    print(f"- {k}: {v}")

# Optional: pretty-print full result when debugging
# import json
# print(json.dumps(result, indent=2))
```

## Notes for production use

- Store API keys in environment variables or a secret manager.
- Send real logs from your app (support/chat sessions) for meaningful evaluations.
- Keep turn ordering correct (`turn_index` monotonic).
- If needed, override endpoint with `PROOFAGENT_BASE_URL`.
- Use BYO LLM when you want explicit control over model/provider selection for LLM-driven reasoning behavior.
