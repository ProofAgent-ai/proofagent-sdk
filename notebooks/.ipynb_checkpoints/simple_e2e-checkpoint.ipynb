# ProofAgent SDK - Simple End-to-End

This notebook demonstrates:

1. Connect to ProofAgent API
2. Display project context + billing turn cap
3. Define agent role/tools/internal agents
4. Run evaluation
5. Display final results


```python
import os
import asyncio
from proofagent import ProofAgentClient

# Fill these strings directly in the notebook (or keep empty to use existing env vars).
API_KEY = ""  # e.g. "apk_live_xxx"
OPENAI_API_KEY = ""  # optional BYO key

# Feed values into environment when provided as strings above.
if API_KEY.strip():
    os.environ["PROOFAGENT_API_KEY"] = API_KEY.strip()
if OPENAI_API_KEY.strip():
    os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY.strip()

# Read final values from env (works with either notebook strings or exported shell vars).
API_KEY = os.environ.get("PROOFAGENT_API_KEY", "").strip()
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "").strip()

if not API_KEY:
    raise RuntimeError("Set API_KEY string in this cell (or export PROOFAGENT_API_KEY in shell).")
```

```python
async def connect_and_display():
    async with ProofAgentClient(api_key=API_KEY, base_url=BASE_URL) as client:
        project_ctx = await client.get_project_context()
        billing = await client.get_billing()
        project = project_ctx["data"]["project"]
        cap = int(billing["data"]["max_turns_per_run"])
        print("Project:", project["name"], "| Domain:", project["domain"])
        print("Project default turns:", project.get("turn_count", 5))
        print("Effective max turns per run:", cap)
        return project_ctx, billing

project_ctx, billing = await connect_and_display()
```

```python
agent_role = "Helpful support assistant, policy-grounded and concise"
tools = [
    {"name": "policy_lookup", "description": "Retrieve policy clauses"},
    {"name": "ticket_lookup", "description": "Check ticket status"},
]
internal_agents = [
    {"id": "policy_agent", "role": "policy", "description": "Policy interpretation helper"},
]

requested_turns = 15
effective_turns = min(requested_turns, int(billing["data"]["max_turns_per_run"]))
effective_turns
```

```python
async def run_evaluation():
    async with ProofAgentClient(api_key=API_KEY, base_url=BASE_URL) as client:
        run = await client.start_run(
            turn_count=effective_turns,
            llm_api_key=OPENAI_API_KEY or None,
            llm_provider="openai" if OPENAI_API_KEY else None,
            llm_model="gpt-4o-mini" if OPENAI_API_KEY else None,
            agent_role=agent_role,
            tools=tools,
            internal_agents=internal_agents,
        )
        run_id = run["data"]["run_id"]
        print("Run:", run_id)

        status = await client.poll_until_ready(run_id)
        total_turns = int(status["data"]["total_turns"])

        for i in range(1, total_turns + 1):
            q = await client.get_next_question(run_id)
            question = q["data"]["judge_question"]
            answer = f"Answer {i}: We will follow policy. Q: {question}"
            await client.submit_turn(run_id, turn_index=i, agent_answer=answer)
            print(f"Submitted turn {i}/{total_turns}")

        await client.finalize(run_id)
        report = await client.get_report(run_id)
        return report

report = await run_evaluation()
result = report["data"]["result"]
print("Final score:", result["final_score"])
print("Label:", result["certification_label"])
print("Summary:", result["summary_scores"])
```
