# ProofAgent × LangGraph (ReAct) — IT Ops / customer-support agent

This notebook wires a **LangGraph ReAct-style agent** (tool-calling loop) so your **real** LLM answers each **judge question** during a **judge-led** evaluation.

## Credentials (same BYO key for both)

| Use | Variable |
|-----|----------|
| ProofAgent API | `PROOFAGENT_API_KEY` (`apk_live_…`) |
| OpenAI (BYO for **both** the judge pipeline and your agent) | `OPENAI_API_KEY` |

Pass the **same** `OPENAI_API_KEY` into `start_run(llm_api_key=…)` and into `ChatOpenAI(api_key=…)` so one quota/account backs both sides.

## Persona

The agent plays an **IT operations / customer support** assistant: tickets, access, incidents, safe guidance—concise and policy-aware.

````python
import os
import sys
import asyncio
from pathlib import Path

# --- ProofAgent SDK (repo layout: add src + examples) ---
_here = Path.cwd().resolve()
for root in (_here, _here.parent):
    if (root / "src" / "proofagent").is_dir():
        sys.path.insert(0, str(root / "src"))
    if (root / "examples" / "report_helpers.py").is_file():
        sys.path.insert(0, str(root / "examples"))
        break

os.environ.setdefault("PROOFAGENT_API_KEY", "")
os.environ.setdefault("OPENAI_API_KEY", "")

# Paste for local runs, or rely on env / Colab secrets
if not os.environ.get("PROOFAGENT_API_KEY", "").strip():
    os.environ["PROOFAGENT_API_KEY"] = ""  # apk_live_…
if not os.environ.get("OPENAI_API_KEY", "").strip():
    os.environ["OPENAI_API_KEY"] = ""  # sk-…

assert os.environ["PROOFAGENT_API_KEY"].strip(), "Set PROOFAGENT_API_KEY"
assert os.environ["OPENAI_API_KEY"].strip(), "Set OPENAI_API_KEY (BYO for judge + agent)"

OPENAI_MODEL = os.environ.get("OPENAI_MODEL", "gpt-4o-mini")
````

````python
import os

from langchain_core.messages import AIMessage
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent
from langchain_core.tools import tool

from proofagent import ProofAgentClient
from report_helpers import print_full_evaluation_report, print_live_turn_banner

ITOPS_SYSTEM_PROMPT = """You are an IT operations and customer support assistant for an enterprise.
You help with tickets, access requests, incidents, outages, and policy-safe guidance.
Be concise, accurate, and security-conscious. If the user (here: the evaluation judge) asks something
outside IT/support scope, answer helpfully but do not invent internal system data."""


@tool
def ticket_status_stub(ticket_id: str) -> str:
    """Simulated ticket lookup — replace with your ITSM API in production."""
    return f"[stub] Ticket {ticket_id}: acknowledged; replace this tool with real integration."


def build_itops_react_agent():
    llm = ChatOpenAI(
        model=OPENAI_MODEL,
        api_key=os.environ["OPENAI_API_KEY"],
        temperature=0.2,
    )
    # ReAct agent: model may call tools then answer
    graph = create_react_agent(
        llm,
        tools=[ticket_status_stub],
        prompt=ITOPS_SYSTEM_PROMPT,
    )
    return graph


async def agent_reply_to_judge(graph, judge_question: str) -> str:
    """Run the LangGraph agent on the judge's question; return final text for submit_turn."""
    out = await graph.ainvoke(
        {"messages": [{"role": "user", "content": judge_question}]},
        config={"recursion_limit": 25},
    )
    msgs = out.get("messages") or []
    if not msgs:
        return ""
    last = msgs[-1]
    if isinstance(last, AIMessage):
        return (last.content or "").strip()
    return str(getattr(last, "content", last)).strip()
````

````python
async def run_judge_led_with_langgraph_agent():
    graph = build_itops_react_agent()
    byo = os.environ["OPENAI_API_KEY"].strip()

    async with ProofAgentClient.from_env() as client:
        ctx = await client.get_project_context()
        bill = await client.get_billing()
        proj = ctx["data"]["project"]
        cap = int(bill["data"]["max_turns_per_run"])
        turns = min(int(proj.get("turn_count", 5)), cap)

        run = await client.start_run(
            turn_count=turns,
            llm_api_key=byo,
            llm_provider="openai",
            llm_model=OPENAI_MODEL,
            agent_role=(
                "IT operations and customer support agent: incidents, access, tickets, "
                "and safe user-facing guidance"
            ),
            tools=[
                {"name": "ticket_status_stub", "description": "Look up ticket status (demo stub)"},
                {"name": "knowledge_base", "description": "Internal KB search (declared for evaluation)"},
            ],
            internal_agents=[
                {
                    "id": "itops_escalation",
                    "role": "escalation",
                    "description": "Routes severe incidents to on-call",
                },
            ],
        )
        rid = run["data"]["run_id"]
        print("run_id:", rid)

        await client.poll_until_ready(rid)
        st = await client.get_run_status(rid)
        total = int(st["data"]["total_turns"])
        print(f"Interactive phase: {total} turn(s) — LangGraph agent answers each judge question.")

        for i in range(1, total + 1):
            print_live_turn_banner(i, total, phase="get_next_question")
            q = await client.get_next_question(rid)
            jq = q["data"]["judge_question"]
            print_live_turn_banner(i, total, phase="langgraph_react")
            answer = await agent_reply_to_judge(graph, jq)
            print_live_turn_banner(i, total, phase="submit_turn")
            await client.submit_turn(rid, turn_index=i, agent_answer=answer)
            preview = (answer[:280] + "…") if len(answer) > 280 else answer
            print(f"  ✓ Turn {i}/{total} | answer preview: {preview!r}")

        await client.finalize(rid)
        report = await client.get_report(rid)
        print_full_evaluation_report(report)
        return report


# Jupyter: top-level await (use asyncio.run(...) in a plain .py script)
report = await run_judge_led_with_langgraph_agent()
````
